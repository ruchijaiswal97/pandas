```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Number of clusters
num_clusters = 60

# Number of data points per cluster
num_points_per_cluster = 100

# Generate random data for each cluster
data = []
for i in range(num_clusters):
    cluster_center = np.random.rand(2) * 10  # Random cluster center
    cluster_data = cluster_center + np.random.randn(num_points_per_cluster, 2)  # Random data around cluster center
    data.append(cluster_data)

# Flatten the list of arrays into a single array
data = np.concatenate(data)

# Create cluster labels
cluster_labels = np.repeat(np.arange(num_clusters), num_points_per_cluster)

# Create DataFrame
df = pd.DataFrame(data, columns=['X', 'Y'])
df['Cluster'] = cluster_labels

# Plot scatter plot
plt.figure(figsize=(10, 8))
for cluster_id in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster_id]
    plt.scatter(cluster_data['X'], cluster_data['Y'], label=f'Cluster {cluster_id}', alpha=0.5)

plt.title('Scatter Plot with Multiple Clusters')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
